# Comparison analysis: 
    Compare the number of deaths across different regions, years, or causes of death. 
    This can help identify variations and trends in the data

# 1.Compare Total Deaths by Region: 
    Calculate the total number of deaths for each region and compare them using a bar chart or a box plot. 
    This will help you understand if there are significant differences in the total deaths among different regions.

```python
# Group the data by region and calculate the sum of total deaths
total_deaths_by_region = mcd.groupby('Region')['Total Death'].sum().sort_values(ascending=False)
total_deaths_by_region
```

```python
# Plot the total deaths by region
plt.figure(figsize=(10, 6))
total_deaths_by_region.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Total Deaths')
plt.title('Total Deaths by Region')
plt.xticks(rotation=45)
plt.show()
```

# 2.Compare Total Deaths by Year: 
    Analyze the total number of deaths over the years. You can plot a line chart or a bar chart to observe any trends or patterns in the data. This will help you identify if there are any significant changes in the total deaths from year to year

```python
# Group the data by year and calculate the sum of total deaths
total_deaths_by_year = mcd.groupby('Year')['Total Death'].sum()
total_deaths_by_year
```

```python
# Plot the total deaths by year
plt.figure(figsize=(10, 6))
total_deaths_by_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.title('Total Deaths by Year')
plt.xticks(total_deaths_by_year.index)
plt.grid(True)
plt.show()
```

# 3.Compare Causes of Death: 
    Explore the different causes of death and compare their frequencies. You can create a bar chart to visualize the number of deaths from different causes. This analysis will provide insights into which causes contribute the most to the total deaths.

```python
# Calculate the sum of deaths for each cause of death
cause_of_death_counts = mcd.drop(['Region', 'Year', 'Month', 'Total Death'], axis=1).sum()
cause_of_death_counts
```

```python
# Sort the causes of death in descending order of frequencies
cause_of_death_counts_sorted = cause_of_death_counts.sort_values(ascending=False)
cause_of_death_counts_sorted
```

```python
# Plot the frequencies of different causes of death
plt.figure(figsize=(10, 6))
cause_of_death_counts_sorted.plot(kind='bar')
plt.xlabel('Cause of Death')
plt.ylabel('Frequency')
plt.title('Frequencies of Causes of Death')
plt.xticks(rotation=45)
plt.show()
```

# 4.Compare Motor Vehicle Accidents with Other Causes:
    Calculate the sum of deaths for 'Motor Vehicle Accidents' and compare it with the sum of deaths for other causes.
    Create a bar chart or a pie chart to visualize the comparison between motor vehicle accidents and other causes of death.

```python
# Convert the 'Motor Vehicle Accidents' column to numeric
mcd['Motor Vehicle Accidents'] = pd.to_numeric(mcd['Motor Vehicle Accidents'], errors='coerce')

# Calculate the sum of deaths for 'Motor Vehicle Accidents' and other causes
motor_accident_deaths = mcd['Motor Vehicle Accidents'].sum()
other_causes_deaths = mcd.select_dtypes(include='number').sum().sum()

print("Motor Vehicle Accidents Deaths:", motor_accident_deaths)
print("Other Causes Deaths:", other_causes_deaths)
```

```python
# Create a DataFrame for the comparison
comparison_data = pd.DataFrame({'Cause of Death': ['Motor Vehicle Accidents', 'Other Causes'],
                               'Total Deaths': [motor_accident_deaths, other_causes_deaths]})
```

```python
# Create a bar chart to visualize the comparison
plt.figure(figsize=(8, 6))
plt.bar(comparison_data['Cause of Death'], comparison_data['Total Deaths'])
plt.xlabel('Cause of Death')
plt.ylabel('Total Deaths')
plt.title('Comparison of Motor Vehicle Accidents with Other Causes of Death')
plt.grid(True)
plt.show()
```

```python
# Create a pie chart to visualize the comparison
plt.figure(figsize=(8, 6))
plt.pie(comparison_data['Total Deaths'], labels=comparison_data['Cause of Death'], autopct='%1.1f%%')
plt.title('Percentage of Motor Vehicle Accidents vs Other Causes of Death')
plt.show()
```

#  4.Compare Deaths by Month:
    Group the data by the 'Month' column and calculate the sum of deaths for each month.
    Plot a line chart or a bar chart to observe the distribution of deaths across different months.

```python
# Group the data by 'Month' and calculate the sum of deaths
deaths_by_month = mcd.groupby('Month')['Total Death'].sum().reset_index()
deaths_by_month
```

```python
# Plot a bar chart to visualize the distribution of deaths by month
plt.figure(figsize=(10, 6))
plt.bar(deaths_by_month['Month'], deaths_by_month['Total Death'])
plt.xlabel('Month')
plt.ylabel('Total Deaths')
plt.title('Distribution of Deaths by Month')
plt.grid(True)
plt.show()
```

# Outlier detection:

```python
# Select the numeric columns for outlier detection
numeric_columns = ['Total Death', 'Natural Death', 'Septicemia Deaths', 'Deaths from Cancer', 'Diabetes Deaths',
                   'Deaths from Disorder', 'Deaths from Infections', 'Deaths from COPD and Asthma',
                   'Deaths from CLR Diseases', 'Deaths from kidney Diseases', 'Deaths from Symptoms Signs',
                   'Deaths from Heart Diseases', 'Deaths from Brain Diseases', 'Deaths from Accidents',
                   'Motor Vehicle Accidents', 'Deaths from Suicide', 'Deaths from Homicide', 'Drug Overdose']
```

# Data distributions: 
    Examine the distributions of variables using histograms or density plots. This can help identify skewness, multimodality, or any other distributional characteristics in the data.

```python
# Plot histograms for each numeric column
plt.figure(figsize=(12, 8))
for column in numeric_columns:
    plt.hist(mcd[column], bins=20, alpha=0.5, label=column)

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histograms of Numeric Variables')
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Plot density plots for each numeric column
plt.figure(figsize=(12, 8))
for column in numeric_columns:
    sns.kdeplot(mcd[column], label=column)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Plots of Numeric Variables')
plt.legend()
plt.show()
```

```python
from scipy.stats import ttest_ind

# Select the variables for comparison
deaths_cancer = mcd['Deaths from Cancer']
deaths_accidents = mcd['Motor Vehicle Accidents']

# Perform independent t-test
t_statistic, p_value = ttest_ind(deaths_cancer, deaths_accidents)

# Print the results
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
```
